```python
import time
import transitions
from transitions.extensions import GraphMachine
import matplotlib.pyplot as plt
from IPython.display import clear_output
import random

def draw_fsm(machine):
    """Draws and updates the state machine visualization."""
    graph = machine.get_graph()
    graph.draw("fsm.png", prog="dot")  # Save the graph
    img = plt.imread("fsm.png")
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis("off")
    clear_output(wait=True)
    plt.show()

# Create FSM class
class FSM():
    pass
```

```python
states = ["PUSH", "+1", "-1", "PULL", "-ERR", "+ERR"]
transitions = [
    {"trigger": "next", "source": "PULL", "dest": "+1"},
    {"trigger": "next", "source": "+1", "dest": "PUSH"},
    {"trigger": "next_err", "source": "+1", "dest": "+ERR"},
    {"trigger": "next", "source": "+ERR", "dest": "+1"},
    {"trigger": "next", "source": "PUSH", "dest": "-1"},
    {"trigger": "next", "source": "-1", "dest": "PULL"},
    {"trigger": "next_err", "source": "-1", "dest": "-ERR"},
    {"trigger": "next", "source": "-ERR", "dest": "-1"},
]

fsm = FSM()
machine = GraphMachine(model=fsm, states=states, transitions=transitions, initial='PULL')

# Step through states every second
end = False
try:
    while not end:
        fsm.trigger('next')
        draw_fsm(machine)
        time.sleep(0.1)
except KeyboardInterrupt:
    print("ending simulation")
    end = True
```

```python
# M<2
states = ["PULL", "+1", "PUSH", "ERR"]
transitions = [
    {"trigger": "next1", "source": "PULL", "dest": "+1"},
    {"trigger": "next2", "source": "+1", "dest": "PUSH"},
    {"trigger": "next3", "source": "PUSH", "dest": "+1"},
    {"trigger": "next4", "source": "+1", "dest": "PULL"},
    {"trigger": "next_err", "source": "+1", "dest": "ERR"},
    {"trigger": "next", "source": "ERR", "dest": "+1"},
]
ERROR_PROB = 0.25

fsm = FSM()
machine = GraphMachine(model=fsm, states=states, transitions=transitions, initial='PULL')

# Step through states every second
end = False
try:
    i = 0
    error = False
    while not end:
        match (i%4):
            case 0:
                error = random.random() < ERROR_PROB
                fsm.trigger('next1')
            case 1:
                if (error):
                    fsm.trigger('next_err')
                else:
                    fsm.trigger('next2')
            case 2:
                if (error):
                    fsm.trigger('next')
                else:
                    fsm.trigger('next3')
            case 3:
                fsm.trigger('next4')


        i+=1
        draw_fsm(machine)
        time.sleep(0.5)
except KeyboardInterrupt:
    print("ending simulation")
    end = True
```

```python
# DAE
states = ["EQBM", "+1", "+2", "+3", "ERR+1", "ERR+2", "ERR+3"]
transitions = [
    {"trigger": "send", "source": "EQBM", "dest": "+1"},
    {"trigger": "ack", "source": "+1", "dest": "+2"},
    {"trigger": "send", "source": "+2", "dest": "+3"},
    {"trigger": "ack", "source": "+3", "dest": "EQBM"},
    {"trigger": "err", "source": "+1", "dest": "ERR+1"},
    {"trigger": "err", "source": "+2", "dest": "ERR+2"},
    {"trigger": "err", "source": "+3", "dest": "ERR+3"},
    {"trigger": "next", "source": "ERR+1", "dest": "EQBM"},
    {"trigger": "next", "source": "ERR+2", "dest": "ERR+1"},
    {"trigger": "next", "source": "ERR+3", "dest": "ERR+2"},
]
ERROR_PROB = 0.1

fsm = FSM()
machine = GraphMachine(model=fsm, states=states, transitions=transitions, initial='EQBM')

# Step through states every second
end = False
try:
    error = False
    while not end:
        match (fsm.state):
            case "EQBM":
                fsm.trigger('send')
            case "+1":
                if (random.random() < ERROR_PROB):
                    fsm.trigger('err')
                else:
                    fsm.trigger('ack')
            case "+2":
                if (random.random() < ERROR_PROB):
                    fsm.trigger('err')
                else:
                    fsm.trigger('send')
            case "+3":
                if (random.random() < ERROR_PROB):
                    fsm.trigger('err')
                else:
                    fsm.trigger('ack')
            case "ERR+1":
                fsm.trigger('next')
            case "ERR+2":
                fsm.trigger('next')
            case "ERR+3":
                fsm.trigger('next')
        draw_fsm(machine)
        time.sleep(0.5)
except KeyboardInterrupt:
    print("ending simulation")
    end = True
```
